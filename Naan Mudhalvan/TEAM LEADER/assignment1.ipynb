5.handling missing values


```python
df.isnull()
```

```python
df.isnull().sum()

```

```python
#filling null values
df.fillna(method="pad")

df.fillna(value="0")
df.fillna(method="bfill")
df.fillna(value=df['postalcode'].mean())
#deleting null values
df.dropnull(how='any')
#replace null values
import numpy as np
df.replace(replace_to=np.nan,values='3')
#interpolate
df["built year"]=df["builtyear"].interpolate(method="linear")
```
